## Constant Time Complexity

In constant time complexity, the execution time of the algorithm remains the same regardless of the input size. It is denoted as O(1).

```python
def average(values):
    average = 0
    for value in values:
        average += value
    return average / len(values)
coefficients = [2,2]
order = "O(N)"
print(order)
```

### Time Complexity based on complexity of functional calls

```python
def sum_values(values):
    total = 0            # 1
    for value in values: # N
        total += value   # N
    return total         # 1

def num_values(values):
    total = 0            # 1
    for _ in values:     # N
        total += 1       # N
    return total         # 1

def average(values):
    value_sum = sum_values(values)  # N
    num_values = num_values(values) # N
    return value_sum / num_values   # 1
```

### List Complexity
append is faster than insert in list

```python
def add_with_append(N):
    values = []
    for i in range(N):
        values.append(i)
    return values

def add_with_insert(N):
    values = []
    for i in range(N):
        values.insert(0, i)
    return values

import time as t 
start = t.time()
add_with_append(50000)
end = t.time()
time_append = round(end - start,2) 
print(time_append)
start = t.time()
add_with_insert(50000)
end = t.time()
time_insert = round(end - start,2) 
print(time_insert)
```

### Time Complexity - String concatenation

```python
import time
import random
import string

def generate_random_strings_list(list_size, string_length):
    characters = string.ascii_letters + string.digits
    
   
    random_strings = [''.join(random.choice(characters) for _ in range(string_length)) for _ in range(list_size)]
    
    return random_strings


random_strings = generate_random_strings_list(10, 8)
 
def concat_with_add(word_list):
    concat = ''
    for word in word_list:
        concat += word
    return concat

def concat_with_join(word_list):
    return ''.join(word_list)

print(random_strings[:5])

start = time.time()
concat_with_add(random_strings)
end = time.time()
time_add = round(end - start,2) 


start = time.time() 
concat_with_join(random_strings)
end = time.time()
time_join = round(end - start,2)


print(time_add)
print(time_join)
```
